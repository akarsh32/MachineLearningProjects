```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
df = pd.read_csv('wine_dataUC.csv')
```

```python
df.head()
#feature_names': ['Class', 
 # 'alcohol',
 # 'malic_acid',
 # 'ash',
 # 'alcalinity_of_ash',
 # 'magnesium',
 # 'total_phenols',
 # 'flavanoids',
 # 'nonflavanoid_phenols',
 # 'proanthocyanins',
 # 'color_intensity',
 # 'hue',
 # 'od280/od315_of_diluted_wines',
 # 'proline']
```

```python
df.columns = ['Class', 'alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
  'proanthocyanins', 'color_intensity', 'hue', 'od280/od315_of_diluted_wines', 'proline']
```

```python
df.head()
```

```python
df.describe().transpose()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df['magnesium'].fillna(df['magnesium'].mean(), inplace=True)
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df['alcalinity_of_ash'].unique()
```

```python
df['alcalinity_of_ash'] = df['alcalinity_of_ash'].apply(lambda aoa:np.nan if (pd.isna(aoa) or pd.isnull(aoa) or aoa==' ') else aoa)
```

```python
df['alcalinity_of_ash'].unique()
df['alcalinity_of_ash'] = df['alcalinity_of_ash'].astype('float64')
df['alcalinity_of_ash'].fillna(df['alcalinity_of_ash'].mean(), inplace=True)
df.info()
```

```python
df['color_intensity'].unique()
```

```python
df['color_intensity'] = df['color_intensity'].apply(lambda ci:np.nan if ci=='lost' or ci=='missing' else ci)
```

```python
df['color_intensity'].unique()
df['color_intensity'] = df['color_intensity'].astype('float64')
df['color_intensity'].fillna(df['color_intensity'].mean(), inplace=True)
df.info()
```

```python
sns.pairplot(df, hue='Class')
```

```python
plt.figure(figsize=[10, 5])
sns.heatmap(df.corr(), annot=True)
```

```python
df['Class'].value_counts()
```

```python
x = df.iloc[:, 1:]
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lrm = LogisticRegression()
```

```python
lrm.fit(X_train, y_train)
predictions = lrm.predict(X_test)
print('Logistic Training Data Score = ', lrm.score(X_train, y_train))
print('Logistic Testing Data Score = ', lrm.score(X_test, y_test))
```

```python
#check with svm
```

```python
from sklearn.svm import SVC
svclassifier = SVC()
```

```python
svclassifier.fit(X_train, y_train)
predictions = svclassifier.predict(X_test)
print('SVM Training Data Score = ', svclassifier.score(X_train, y_train))
print('SVM Testing Data Score = ', svclassifier.score(X_test, y_test))
```

```python
# now fit svm with different gamma and c values using grid search
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1, 1, 10, 100, 1000], 'gamma':[1,0.1,0.01, 0.001, 0.0001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
# so grisearch predicted on best values of c and gamma possible
```

```python
gridPredictions = grid.predict(X_test)
print('SVM grid Training Data Score = ', grid.score(X_train, y_train))
print('SVM grid Testing Data Score = ', grid.score(X_test, y_test))
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, gridPredictions))
```

```python
print(classification_report(y_test, gridPredictions))
```
